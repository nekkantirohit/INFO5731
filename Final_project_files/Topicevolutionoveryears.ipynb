```python
import pandas as pd
df=pd.read_csv('/content/sample_data/dynamic topics.csv')
df.head()
```

```python
df['year'] = pd.DatetimeIndex(df['Timestamp']).year
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.subplots(figsize=(10,10))
sns.barplot(data=df,x= 'year',y= 'Frequency',hue= 'Topic')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left',title='Topic', borderaxespad=0)
#sns.barplot(,pos=below)
plt.xlabel("year", size=20)
plt.ylabel("frequency of the respective topic", size=20)
plt.title("Representation of the topics over the years with high frequency", size=24)
```
