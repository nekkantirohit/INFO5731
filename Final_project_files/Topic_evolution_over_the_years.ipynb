```python
import pandas as pd
df=pd.read_csv('/content/sample_data/dynamic topics.csv')
df.head()
```

```python
df['year'] = pd.DatetimeIndex(df['Timestamp']).year
```

```python
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.subplots(figsize=(18,18))
sns.barplot(y='Topic',x='year',data=df)
```

```python
plt.subplots(figsize=(10,10))
sns.lineplot(y='Topic',x='year',data=df,)
plt.xlabel("year", size=20)
plt.ylabel("Topics", size=20)
plt.title("Topic Evolution over the years ", size=24)
```
