```python
# Write your code here
#importing libraries
import nltk
import pandas as pd
#reading csv file
dafrme = pd.read_csv('shangchi_reviews.csv')
#determining the columns
dafrme['word_count'] = dafrme['Review'].apply(lambda x:len(str(x).split(" ")))
dafrme
```

```python
#removal of punctuations
dafrme['Review'] = dafrme['Review'].str.replace('[^\w\s]','') 
#converting the text to lower case
dafrme['Review'] = dafrme['Review'].apply(lambda x: " ".join(x.lower() for x in x.split()))
#removal of stopwords
#importing libraries
from nltk.corpus import stopwords
nltk.download('stopwords')
stop = stopwords.words('english')
dafrme['Review'] = dafrme['Review'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
```

```python
#implementing the text in single line
single_line = ''
#loop for determning  the length
for i in range(0,len(dafrme['Review'])):
  single_line+= dafrme['Review'][i]
print(single_line)
```

```python
#(1) Count the frequency of all the N-grams (N=3).
#trigrams
#importing the modules
nltk.download('punkt')
words = nltk.word_tokenize(single_line)
res_trigrams = nltk.trigrams(words)
#applying FreqDist using nltk module
freq_trigrams = nltk.FreqDist(res_trigrams)
freq_trigrams
```

(2) Calculate the probabilities for all the bigrams in the dataset by using the fomular count(w2 w1) / count(w2). For example, count(really like) / count(really) = 1 / 3 = 0.33.

```python
#bigrams
res_bigrams = nltk.bigrams(words)
#importing libraries
import re
from itertools import tee, islice
import collections
#logic for caluculating ngrams
def ngrams(lst, no):
  t_list = lst
  while True:
    x, y = tee(t_list)
    lne = tuple(islice(x, no))
    if len(lne) == no:
      yield lne
      next(y)
      t_list = y
    else:
      break
line_2 = collections.Counter(ngrams(single_line.split(), 2))
print(line_2)
```
