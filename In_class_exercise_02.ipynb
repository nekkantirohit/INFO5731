```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
from time import sleep
from random import randint
```

```python
headers = {"Accept-Language": "en-US,en;q=0.5"}
```

```python
url = 'https://www.imdb.com/title/tt6751668/reviews?ref_=tt_urv'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
Account_name = []
Star = []
Review_heading = []
Review_text = []
Review_upload_time = []
```

```python
review_data = soup.findAll('div', attrs= {'class': "text show-more__control"})
```

```python

review_data 
```

```python
name = soup.findAll('span', class_='display-name-link')
```

```python
name
```

```python
account_names = []
for i in range(0,len(name)):
    account_names.append(name[i].get_text())
```

```python
title = soup.findAll('a', class_='title')
```

```python
Review_heading = []
for i in range(0,len(title)):
    Review_heading.append(title[i].get_text())
```

```python
Review_texts = soup.findAll('div', class_='text show-more__control')
```

```python
Review_text = []
for i in range(0,len(Review_texts)): 
    Review_text.append(Review_texts[i].get_text())
```

```python
posted_time = soup.findAll('span', class_='review-date')
```

```python
Review_upload_time = []
for i in range(0,len(posted_time)): 
    Review_upload_time.append(posted_time[i].get_text())
```

```python
import pandas as pd
scrapping_DF = pd.DataFrame({'Account_name': account_names,'Review_heading': Review_heading, 'Review_text': Review_text, 'Review_upload_time': Review_upload_time})
```
