```python
#importing selenium modules from webdriver module
from selenium import webdriver
#assigning a variable to the object
options = webdriver.ChromeOptions()
#calling objects
options.add_argument('-headless')
options.add_argument('-no-sandbox')
options.add_argument('-disable-dev-shm-usage')
```

```python
from selenium.webdriver.support.ui import WebDriverWait as wait
from selenium.webdriver.common.by import By
import pandas as pd
import time

driver = webdriver.Chrome('chromedriver',options=options)
#assigning link
link = 'https://www.imdb.com/title/tt9376612/reviews?ref_=tt_sa_3'
#intializing arrays
movie_array = []
review_array = []
#calling link using web driver module
driver.get(link)
#defining the range of numbers
for num in range(43):
  #finding load more button class using driver module
  driver.find_element_by_class_name("ipl-load-more__button").click()
  #suspends the excution for certain time
  time.sleep(5)
  #finding title and text classes using driver module
  listOfMovies = driver.find_elements(By.CLASS_NAME, "title")
  listOfReviews = driver.find_elements(By.CLASS_NAME, "text")
#mapping similar index iterators using zip function ina aloop
for ele, sub_ele in zip(listOfMovies, listOfReviews):
      movie_array.append((ele.text).replace('\n',''))
      review_array.append(sub_ele.text)
#defining data frame
dafrme = pd.DataFrame(list(zip(movie_array, review_array)), columns =['Movies', 'Review'])
print("Length of data frame is {0}".format(len(dafrme)))
dafrme
```

```python
dafrme.to_csv('shangchi_review.csv', index=False)
```

```python
#Frequent words removal
frequent = pd.Series(' '.join(dafrme['Review']).split()).value_counts()[:10]
frequent
```

```python
frequent = list(frequent.index)
dafrme['Frequent words'] = dafrme['Review'].apply(lambda x: " ".join(x for x in x.split() if x not in frequent))
dafrme['Review'].head()
```

```python
#Rare words removal
frequent = pd.Series(' '.join(dafrme['Review']).split()).value_counts()[-10:]
frequent
```

```python
#assigning variable
frequent = list(frequent.index)
dafrme['rare words'] = dafrme['Review'].apply(lambda x: " ".join(x for x in x.split() if x not in frequent))
dafrme['Review'].head()
```

```python
#ngrams
import nltk
nltk.download('punkt')
from textblob import TextBlob
TextBlob(dafrme['Review'][0]).ngrams(2)
```

```python
#term frequnecy
tf1 = (dafrme['Review'][0:10]).apply(lambda x: pd.value_counts(x.split(" "))).sum(axis = 0).reset_index()
#assigning colums
tf1.columns = ['words','tf']
tf1
```

```python
import nltk
nltk.download('stopwords')
#importing re module
import re
from nltk.corpus import stopwords
### Dataset Preprocessing
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
test=[]
for i in range(0, len(dafrme)):
    print(i)
    movie = re.sub('[^a-zA-Z]', ' ', dafrme['Review'][i])
    #converting into lowercase
    movie= movie.lower()
    #spliting the string of reviews into list of words
    movie = movie.split()    
    #counting stopwords
    movie = [ps.stem(word) for word in movie if not word in stopwords.words('english')]
    movie = ' '.join(movie)
    test.append(movie)
```

```python
from sklearn.feature_extraction.text import CountVectorizer


#Creating the bag of words
bow = CountVectorizer().fit(test)
text = bow.transform(test)
counts=bow.get_feature_names()


counts
```

```python
#assigning rows and colums through data frame
dafrme=pd.DataFrame(data=text.toarray(),columns=counts)
dafrme
```
