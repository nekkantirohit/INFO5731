```python
#importing selenium modules from webdriver module
from selenium import webdriver
#assigning a variable to the object
options = webdriver.ChromeOptions()
#calling objects
options.add_argument('-headless')
options.add_argument('-no-sandbox')
options.add_argument('-disable-dev-shm-usage')
```

```python
from selenium.webdriver.support.ui import WebDriverWait as wait
from selenium.webdriver.common.by import By
import pandas as pd
import time

driver = webdriver.Chrome('chromedriver',options=options)
#assigning link
link = 'https://www.imdb.com/title/tt9376612/reviews?ref_=tt_sa_3'
#intializing arrays
movie_array = []
review_array = []
#calling link using web driver module
driver.get(link)
#defining the range of numbers
for num in range(43):
  #finding load more button class using driver module
  driver.find_element_by_class_name("ipl-load-more__button").click()
  #suspends the excution for certain time
  time.sleep(5)
  #finding title and text classes using driver module
  listOfMovies = driver.find_elements(By.CLASS_NAME, "title")
  listOfReviews = driver.find_elements(By.CLASS_NAME, "text")
#mapping similar index iterators using zip function ina aloop
for ele, sub_ele in zip(listOfMovies, listOfReviews):
      movie_array.append((ele.text).replace('\n',''))
      review_array.append(sub_ele.text)
#defining data frame
dafrme = pd.DataFrame(list(zip(movie_array, review_array)), columns =['Movies', 'Review'])
print("Length of data frame is {0}".format(len(dafrme)))
dafrme
```

```python
dafrme.to_csv('shangchi_review.csv', index=False)
```

#**(1) Basic feature extraction using text data**

```python

#Number of characters
dafrme['num_char'] = dafrme['Review'].str.len()
dafrme[['Review','num_char']].head()
```

```python
#Number of words
#splitting the string into list of words
dafrme['word_count'] = dafrme['Review'].apply(lambda x: len(str(x).split(" ")))
dafrme[['Review','word_count']].head()
```

```python
#average word length
#defining function
def avg_word_length(sent):
  #splitting the word into list
  words = sent.split()
  return (sum(len(word) for word in words)/(len(words)+1))
#calling dataframe
dafrme['avg_word_length'] = dafrme['Review'].apply(lambda x: avg_word_length(x))
dafrme[['Review','avg_word_length']].head()
```

```python
#Number of stopwords
#importing stopwords from nltk module
from nltk.corpus import stopwords
stop = stopwords.words('english')

dafrme['stopwords'] = dafrme['Review'].apply(lambda x: len([x for x in x.split() if x in stop]))
dafrme[['Review','stopwords']].head()
```

```python
#Number of special characters
dafrme['hastags'] = dafrme['Review'].apply(lambda x: len([x for x in x.split() if x.startswith('#')]))
dafrme[['Review','hastags']].head()
```

```python
#Number of numerics
dafrme['numerics'] = dafrme['Review'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
dafrme[['Review','numerics']].head()
```

```python
#Number of uppercase words
dafrme['upper'] = dafrme['Review'].apply(lambda x: len([x for x in x.split() if x.isupper()]))
dafrme[['Review','upper']].head()
```

#**(2) Basic Text Pre-processing of text data**

```python
#importing nltk module
import nltk
nltk.download('stopwords')
```

```python
#lower case
dafrme['Review'] = dafrme['Review'].apply(lambda x: " ".join(x.lower() for x in x.split()))
dafrme['Review'].head()
```

```python
#remove punctuations and special characters
import re
dafrme['Review']=dafrme['Review'].str.replace('[^\w\s]','')
dafrme['Review']=dafrme['Review'].apply(lambda x: ''.join(re.sub(r"[^a-zA-Z0-9]+", ' ', charactr) for charactr in x ))

dafrme['Review'].head()
```

```python
#remove stopwords
from nltk.corpus import stopwords
stop = stopwords.words('english')
dafrme['Review'] = dafrme['Review'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
dafrme['Review'].head()
```

```python
#Frequent words removal
frequent = pd.Series(' '.join(dafrme['Review']).split()).value_counts()[:10]
frequent
```

```python
#assigning variable
frequent = list(frequent.index)
dafrme['Review'] = dafrme['Review'].apply(lambda x: " ".join(x for x in x.split() if x not in frequent))
dafrme['Review'].head()
```

```python
#Rare words removal
frequent = pd.Series(' '.join(dafrme['Review']).split()).value_counts()[-10:]
frequent
```

```python
#assigning variable
frequent = list(frequent.index)
dafrme['Review'] = dafrme['Review'].apply(lambda x: " ".join(x for x in x.split() if x not in frequent))
dafrme['Review'].head()
```

```python
#tokenization
import nltk
nltk.download('punkt')
from textblob import TextBlob
TextBlob(dafrme['Review'][1]).words
```

```python
#spell correction
from textblob import TextBlob
dafrme['Review'][:5].apply(lambda x: str(TextBlob(x).correct()))
```

```python
#stemming
#importing porterStemmer module
from nltk.stem import PorterStemmer
s1 = PorterStemmer()
dafrme['Review'][:5].apply(lambda x: " ".join([s1.stem(word) for word in x.split()]))
```

```python
#lemmatization
#importing word module from textblob
from textblob import Word
nltk.download('wordnet')
dafrme['Review'] = dafrme['Review'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
dafrme['Review'].head()
```

#**(1) Parts of Speech (POS) Tagging: Tag Parts of Speech of each word in the text, and calculate the total number of N(oun), V(erb), Adj(ective), Adv(erb), respectively.**

```python
#impoting tags from nltk module
from nltk import word_tokenize, pos_tag, pos_tag_sents
dafrme['POSTags'] =  pos_tag_sents( dafrme['Review'].apply(word_tokenize).tolist() )
print(dafrme)
#converts series of posttags in dataframe to lists
p = dafrme['POSTags'].to_list()
p
#assigning variable
nouns = 0
#loop for counting num of nouns
for f in p:
  for x,y in f:
    #compares the value to substring and returns if the value is equal
    if(y.startswith('N')):
       nouns=nouns+1
       #formats the the specified value and inserts it into string
rslt = "nouns:{}".format(nouns)
print(rslt)
#assigning variable
verb = 0
#loop for counting num of verbs
for f in p:
  for x,y in f:
    if(y.startswith('V')):
       verb=verb+1
       #formats the the specified value and inserts it into string
rslt1 = "verb:{}".format(verb)
print(rslt1)
adverb = 0
#loop for counting num of adverb
for f in p:
  for x,y in f:
    if(y.startswith('R')):
       adverb=adverb+1
       #formats the the specified value and inserts it into string
rslt2 = "adverb:{}".format(adverb)
print(rslt2)
#assigning variable
adj = 0
#loop for counting num of adjectives
for f in p:
  for x,y in f:
    if(y.startswith('J')):
       adj=adj+1
       #formats the the specified value and inserts it into string
rslt3 = "adj:{}".format(adj)
print(rslt3)
```
